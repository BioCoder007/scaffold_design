```python
def get_frag_file(fragfile):#directories[0]
    
    col = ['sequence','worstfrag','worst6frags','sum_best_frags','avg_all_frags']
    f = pd.read_csv(fragfile +'/' + 'score.sc.frag_quals',sep='\s' ,header=None,usecols=[1,2,9,11,12],names=col)                               
    f['n'] = f['sequence'].apply(lambda x: (len(x)- 8) )
    f['avg_best_frag'] = f['sum_best_frags']/ f['n'] 

    #f.head()
    return f
```

```python
out = ['sequence',
     'worstfrag',
     'worst6frags',
     'sum_best_frags',
     'avg_all_frags',
     'avg_best_frag']

    
dfs = {}
for idx in range(len(files)):
    print files[idx],directories[idx]
    df = pd.read_csv(files[idx],sep=',',header=(0))
    print len(df.index)
    df.head()
    
    frags = get_frag_file(directories[idx])
    score_dict = pd.merge(left=df, right=frags[out], on='sequence',how='left')
    score_dict.head()

    dfs[idx] = score_dict

alldata = pd.concat(dfs)
alldata.head()
print len(alldata)
```
